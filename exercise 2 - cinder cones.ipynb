```python
# import modules
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
```

```python
# initialize variables
g = 9.8 # m/s
bombs = 50 # need thousands
alpha = np.radians(np.random.randint(1,179,bombs))
angle = alpha
v0 = 25,50,75
v0 = np.random.randint(5,80,bombs)
base = np.zeros(4000)
x = np.linspace(0,4000,4000)
xstart = 2000
ystart = -1
diameter = 0.25 # keep small relative to dx
dx = 1
dz = (np.pi*(diameter**2)/4)/dx

plt.plot(base,'k-')
plt.title('Initial Topography')
plt.ylim([-10,10])
plt.ylabel('elevation (m)')
plt.show()
```

```python
# run program

for l in range(len(v0)):
    for k in range(len(angle)):
        #for i in range(len(x)):
        Uo = np.cos(angle[k])*v0[l]
        Wo = np.sin(angle[k])*v0[l]
            #y = start + ((Wo/Uo)*(x[i]-start)) - ((g/2)*(((x[i]-start)**2)/(Uo**2)))
        y = ystart + ((Wo/Uo)*(x-xstart)) - ((g/2)*(((x-xstart)**2)/(Uo**2)))
        
        ystart = base[2000]
        
        above=np.where(y>base)[0]
        
        if len(above)== 0:
            base[2000] = base[2000]
            #base[2000] = base[2000] + (0.02*dz)
        elif above[-1] > 2000:
            base[above[-1]] = base[above[-1]] + dz
        elif above[-1] < 2000:
            base[above[0]] = base[above[0]] + dz

            
# this loops through 4000 dxs each time instead of checking just the box the bomb landed in...
# WAYYYYY slower and more computation than checking one box
# paper greg sent around has efficient way of accounting for angle of repose and evolution
        for i in range(len(base)):
            beta = np.arctan((base[i]-base[i-1])/dx)
            if beta > np.radians(32):
                base[i] = base[i] - dz
                base[i+1] = base[i+1] + dz
            elif beta <= np.radians(32):
                base[i] = base[i]
        
        #plt.plot(x[above],y[above])

#plt.ylim(-1,300) 
#plt.plot(base,'k-')
#plt.ylabel('height (m)')
#plt.xlabel('distance (m)')
#plt.show()
```
